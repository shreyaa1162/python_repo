```python
units=int(input("enter units :"))
if units <= 100:
  bill = 0
elif units <= 200:
  bill = (units-100) * 5
elif units <= 300:
  bill=(100*5) + (units-200) * 7
else:
  bill = (100*5) + (100*7) + (units - 300) * 10
print("electricity bill=",bill)
```

```python
s = input("Enter a string: ")

n = len(s)

for i in range(n):
   for j in range(n):
     if i == j:
        print(s[i], end="")
     elif i + j == n - 1:
        print(s[i], end="")
   else:
        print("*", end="")
   print()
```
