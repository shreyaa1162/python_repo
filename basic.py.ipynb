```python
distance = float(input("enter distance:"))
waiting = int(input("enter waiting time:"))
if distance <= 100:
  amount = distance * 10
elif distance <= 200:
  amount = 100 * 10 + (distance - 100) * 9
else:
  amount = 100 * 10 + (distance - 200) * 6
waiting_charge = waiting * 5
total = amount + waiting_charge
print("Amount =", amount)
print("Waiting Charge =", waiting_charge)
print("Total Amount =", total)
```
