```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

max_num = max(num1, num2)

while True:
    if max_num % num1 == 0 and max_num % num2 == 0:
        print("LCM =", max_num)
        break
    max_num += 1
```

```python
a = int(input())
b = int(input())

hcf = 1
i = 1

while i <= a and i <= b:
    if a % i == 0 and b % i == 0:
        hcf = i
    i += 1

print("HCF =", hcf)
```

```python
s = input("Enter a string: ")

n = len(s)

for i in range(n):
    for j in range(n):
        if i == j:
            print(s[i], end="")
        elif i + j == n - 1:
            print(s[i], end="")
        else:
            print("*", end="")
    print()
```
