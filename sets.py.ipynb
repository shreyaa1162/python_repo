```python
string = input("Enter a string: ")
new_str = ""

for i in string:
    if 'a' <= i <= 'z':
        new_str += chr(ord(i) - 32)
    elif 'A' <= i <= 'Z':
        new_str += chr(ord(i) + 32)
    else:
        new_str += i

print("Changed string:", new_str)
```
